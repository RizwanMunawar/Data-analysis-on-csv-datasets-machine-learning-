```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```
data = pd.read_csv('/content/tvmarketing.csv')
```

```
data.head()
```

```
len(data)
```

```
data.isnull().sum()
```

```
data.isnull().values
```

```
data['TV'] = data['TV'].astype(int)
```

```
data.head()
```

```
plt.plot(data['TV'],data['Sales'],'r+')
```

```
plt.scatter(data['TV'],data['Sales'])
```

```
sns.countplot(data['Sales'])
```

```
sns.pairplot(data)
```

```
X = data['TV'].values.reshape(-1,1)
y = data['Sales']

r2_score_list = []
mean_squared = []
r_mean_squared = []
```

```
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.25,random_state = 2)
```

```
from sklearn.linear_model import LinearRegression
```

```
lr = LinearRegression()
```

```
lr.fit(X_train,y_train)
```

```
pred = lr.predict(X_test)
```

```
len(pred)
```

```
print(pred)
```

```
from sklearn.metrics import r2_score,mean_squared_error
```

```
print('R2 Score is {}'.format(r2_score(y_test,pred)))
r2_score_list.append(r2_score(y_test,pred))
print('Mean Square Error is {}'.format(mean_squared_error(y_test,pred)))
mean_squared.append(mean_squared_error(y_test,pred))
print('Root mean Square Error is {}'.format(np.sqrt(mean_squared_error(y_test,pred))))
r_mean_squared.append(np.sqrt(mean_squared_error(y_test,pred)))
```

```
from sklearn.svm import SVR
```

```
SVR = SVR()
```

```
SVR.fit(X_train,y_train)
```

```
y_pred = SVR.predict(X_test)
```

```
print('R2 Score is {}'.format(r2_score(y_test,y_pred)))
r2_score_list.append(r2_score(y_test,y_pred))
print('Mean Square Error is {}'.format(mean_squared_error(y_test,y_pred)))
mean_squared.append(mean_squared_error(y_test,y_pred))
print('Root Means Square Error is {}'.format(np.sqrt(mean_squared_error(y_test,y_pred))))
r_mean_squared.append(np.sqrt(mean_squared_error(y_test,y_pred)))
```

```
from sklearn.ensemble import RandomForestRegressor
```

```
RFR = RandomForestRegressor()
```

```
RFR.fit(X_train,y_train)
```

```
y_pre = RFR.predict(X_test)
```

```
print('R2 Score is {}'.format(r2_score(y_test,y_pre)))
r2_score_list.append(r2_score(y_test,y_pre))
print('Mean Square Error is {}'.format(mean_squared_error(y_test,y_pre)))
mean_squared.append(mean_squared_error(y_test,y_pre))
print('Root Mean Sqaured Error is {}'.format(np.sqrt(mean_squared_error(y_test,y_pre))))
r_mean_squared.append(np.sqrt(mean_squared_error(y_test,y_pre)))
```

```
r2_score_list
mean_squared
r_mean_squared
```

```
dataframe = pd.DataFrame(np.array([r2_score_list, mean_squared, r_mean_squared]).T,
                         columns = ['R2_S','MSE','RMSE'],index = ['LR','SVR','RFR'])
```
